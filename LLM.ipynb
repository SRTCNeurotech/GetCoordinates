```python
import itertools
import numpy as np
import pandas as pd
import Linear_method_error as LM
```

```python
# get permutations
lst = list(itertools.permutations(range(11), r=6))
```

```python
# remove repetitions from permutations, get combinations
for i in range(len(lst)):
    lst[i] = list(lst[i])
    lst[i].sort()
    lst[i] = frozenset(lst[i])
    
lst = frozenset(lst)
lst = list(lst)
for i in range(len(lst)):
    lst[i] = list(lst[i])
```

```python
# Load input data in pandas dataframe 
df = pd.read_excel('sample.xlsx')
for i in range(len(df)):
    df.Centers_region[i] = df.Centers_region[i].lower()
    
# scan area resolution value
LM.scan_area_resolution = 3.321 #Âµm/px
```

```python
# First 5 rows of loaded DataFrame
df.head()
```

```python
# IT IS MANDATORY TO RUN THIS CELL BEFORE EACH NEW RUNNING OF THE CYCLE IN THE CELL BELOW
absolute_error_result = np.zeros((1,3))
```

```python
# CYCLE
for i in range(len(lst)):
    # create a list with ID of reference landmarks:
    lst_ctrl = lst[i]
    # create a list with ID of control landmarks:
    lst_test = []
    for k in range(11):
        if k not in lst_ctrl:
            lst_test.append(k)   
    
    # create DataFrames for: reference landmarks -> df_ctrl, control landmarks -> df_test
    df_ctrl = df.iloc[np.array(lst_ctrl)]
    df_test = df.iloc[np.array(lst_test)]
    
    a_test = df_test[df_test['Centers_region'] == 'a'].iloc[:, 0:8]
    a_work = df_ctrl[df_ctrl['Centers_region'] == 'a'].iloc[:, 0:8]
    b_test = df_test[df_test['Centers_region'] == 'b'].iloc[:, 0:8]
    b_work = df_ctrl[df_ctrl['Centers_region'] == 'b'].iloc[:, 0:8]
    c_test = df_test[df_test['Centers_region'] == 'c'].iloc[:, 0:8]
    c_work = df_ctrl[df_ctrl['Centers_region'] == 'c'].iloc[:, 0:8]
    d_test = df_test[df_test['Centers_region'] == 'd'].iloc[:, 0:8]
    d_work = df_ctrl[df_ctrl['Centers_region'] == 'd'].iloc[:, 0:8]
    e_test = df_test[df_test['Centers_region'] == 'e'].iloc[:, 0:8]
    e_work = df_ctrl[df_ctrl['Centers_region'] == 'e'].iloc[:, 0:8]
    f_test = df_test[df_test['Centers_region'] == 'f'].iloc[:, 0:8]
    f_work = df_ctrl[df_ctrl['Centers_region'] == 'f'].iloc[:, 0:8]
    g_test = df_test[df_test['Centers_region'] == 'g'].iloc[:, 0:8]
    g_work = df_ctrl[df_ctrl['Centers_region'] == 'g'].iloc[:, 0:8]
    
    if len(a_work) > 0:
        repers_work_a = np.array(a_work.values[:, 3:], dtype=float)
        repers_work_a = np.column_stack((repers_work_a, LM.get_stage_coord_mm(repers_work_a[:,1:], 
                                                                              a_work.values[0, 1],
                                                                              a_work.values[0, 2])))
    if len(a_test) > 0:
        repers_test_a = np.array(a_test.values[:, 3:], dtype=float)
        repers_test_a = np.column_stack((repers_test_a, LM.get_stage_coord_mm(repers_test_a[:,1:],
                                                                              a_test.values[0, 1],
                                                                              a_test.values[0, 2])))
    if len(b_work) > 0:
        repers_work_b = np.array(b_work.values[:, 3:], dtype=float)
        repers_work_b = np.column_stack((repers_work_b, LM.get_stage_coord_mm(repers_work_b[:,1:], 
                                                                              b_work.values[0, 1],
                                                                              b_work.values[0, 2])))
    if len(b_test) > 0:
        repers_test_b = np.array(b_test.values[:, 3:], dtype=float)
        repers_test_b = np.column_stack((repers_test_b, LM.get_stage_coord_mm(repers_test_b[:,1:],
                                                                              b_test.values[0, 1],
                                                                              b_test.values[0, 2])))
    if len(c_work) > 0:
        repers_work_c = np.array(c_work.values[:, 3:], dtype=float)
        repers_work_c = np.column_stack((repers_work_c, LM.get_stage_coord_mm(repers_work_c[:,1:], 
                                                                              c_work.values[0, 1],
                                                                              c_work.values[0, 2])))
    if len(c_test) > 0:
        repers_test_c = np.array(c_test.values[:, 3:], dtype=float)
        repers_test_c = np.column_stack((repers_test_c, LM.get_stage_coord_mm(repers_test_c[:,1:],
                                                                              c_test.values[0, 1],
                                                                              c_test.values[0, 2])))
    if len(d_work) > 0:
        repers_work_d = np.array(d_work.values[:, 3:], dtype=float)
        repers_work_d = np.column_stack((repers_work_d, LM.get_stage_coord_mm(repers_work_d[:,1:], 
                                                                              d_work.values[0, 1],
                                                                              d_work.values[0, 2])))
    if len(d_test) > 0:
        repers_test_d = np.array(d_test.values[:, 3:], dtype=float)
        repers_test_d = np.column_stack((repers_test_d, LM.get_stage_coord_mm(repers_test_d[:,1:],
                                                                              d_test.values[0, 1],
                                                                              d_test.values[0, 2])))
    if len(e_work) > 0:
        repers_work_e = np.array(e_work.values[:, 3:], dtype=float)
        repers_work_e = np.column_stack((repers_work_e, LM.get_stage_coord_mm(repers_work_e[:,1:], 
                                                                              e_work.values[0, 1],
                                                                              e_work.values[0, 2])))
    if len(e_test) > 0:
        repers_test_e = np.array(e_test.values[:, 3:], dtype=float)
        repers_test_e = np.column_stack((repers_test_e, LM.get_stage_coord_mm(repers_test_e[:,1:],
                                                                              e_test.values[0, 1],
                                                                              e_test.values[0, 2])))
    if len(f_work) > 0:
        repers_work_f = np.array(f_work.values[:, 3:], dtype=float)
        repers_work_f = np.column_stack((repers_work_f, LM.get_stage_coord_mm(repers_work_f[:,1:], 
                                                                              f_work.values[0, 1],
                                                                              f_work.values[0, 2])))
    if len(f_test) > 0:
        repers_test_f = np.array(f_test.values[:, 3:], dtype=float)
        repers_test_f = np.column_stack((repers_test_f, LM.get_stage_coord_mm(repers_test_f[:,1:],
                                                                              f_test.values[0, 1],
                                                                              f_test.values[0, 2])))
    if len(g_work) > 0:
        repers_work_g = np.array(g_work.values[:, 3:], dtype=float)
        repers_work_g = np.column_stack((repers_work_g, LM.get_stage_coord_mm(repers_work_g[:,1:], 
                                                                              g_work.values[0, 1],
                                                                              g_work.values[0, 2])))
    if len(g_test) > 0:
        repers_test_g = np.array(g_test.values[:, 3:], dtype=float)
        repers_test_g = np.column_stack((repers_test_g, LM.get_stage_coord_mm(repers_test_g[:,1:],
                                                                              g_test.values[0, 1],
                                                                              g_test.values[0, 2])))
    rep_test = np.zeros((1,7))
    if len(a_test) > 0:
        rep_test = np.concatenate((rep_test, repers_test_a))
    if len(b_test) > 0:
        rep_test = np.concatenate((rep_test, repers_test_b))
    if len(c_test) > 0:
        rep_test = np.concatenate((rep_test, repers_test_c))
    if len(d_test) > 0:
        rep_test = np.concatenate((rep_test, repers_test_d))
    if len(e_test) > 0:
        rep_test = np.concatenate((rep_test, repers_test_e))
    if len(f_test) > 0:
        rep_test = np.concatenate((rep_test, repers_test_f))
    if len(g_test) > 0:
        rep_test = np.concatenate((rep_test, repers_test_g))
    rep_test = rep_test[1:,:]
    
    rep_work = np.zeros((1,7))
    if len(a_work) > 0:
        rep_work = np.concatenate((rep_work, repers_work_a))
    if len(b_work) > 0:
        rep_work = np.concatenate((rep_work, repers_work_b))
    if len(c_work) > 0:
        rep_work = np.concatenate((rep_work, repers_work_c))
    if len(d_work) > 0:
        rep_work = np.concatenate((rep_work, repers_work_d))
    if len(e_work) > 0:
        rep_work = np.concatenate((rep_work, repers_work_e))
    if len(f_work) > 0:
        rep_work = np.concatenate((rep_work, repers_work_f))
    if len(g_work) > 0:
        rep_work = np.concatenate((rep_work, repers_work_g))
    rep_work = rep_work[1:,:]
    
    res_rep_dist = LM.calc_min_dist(rep_test, rep_work)
    res_rep = np.column_stack((rep_test, LM.get_anatom_test_coord(res_rep_dist[:,1:], rep_test[:,1:])))
    absolute_error_result = np.concatenate((absolute_error_result, LM.get_absolute_error(res_rep)))
    
    if len(a_work) > 0:
        del repers_work_a
    if len(a_test) > 0:
        del repers_test_a
    if len(b_work) > 0:
        del repers_work_b
    if len(b_test) > 0:
        del repers_test_b
    if len(c_work) > 0:
        del repers_work_c
    if len(c_test) > 0:
        del repers_test_c
    if len(d_work) > 0:
        del repers_work_d
    if len(d_test) > 0:
        del repers_test_d
    if len(e_work) > 0:
        del repers_work_e
    if len(e_test) > 0:
        del repers_test_e
    if len(f_work) > 0:
        del repers_work_f
    if len(f_test) > 0:
        del repers_test_f
    if len(g_work) > 0:
        del repers_work_g
    if len(g_test) > 0:
        del repers_test_g
```

```python
absolute_error_result = absolute_error_result[1:,:]
error_df = pd.DataFrame(data=absolute_error_result, columns=['id_point', 'X', 'Y'])
```

```python
error_df['Error'] = (error_df.X ** 2 + error_df.Y ** 2)**0.5
```

```python
# resulting DataFrame with errors of calculations (first 5 rows)
error_df.head()
```

```python
# save results of calculations of errors into a file 
error_df.to_excel('LLM_error.xlsx', index=False)
```

```python
# Calculation of the percentage of errors that fall within the specified interval.
mas = np.array([])
for i in np.arange(0.02, 0.16, 0.01):
    i = np.round(i, decimals=2)
    mas = np.append(mas, [i*1000, np.round(error_df[error_df['Error'] < i].count()[0] / error_df.count()[0], decimals=2)])
mas = np.reshape(mas, (-1, 2))
```

```python
# save results from previous cell 
mas_df = pd.DataFrame(data=mas, columns=['Error, um', 'Percent, %'])
mas_df.to_excel('LLM_Percent.xlsx', index=False)
```
