```python
import itertools
import numpy as np
import pandas as pd
import Linear_method_error as LM
```

```python
# get permutations
lst = list(itertools.permutations(range(11), r=6))
# remove repetitions from permutations, get combinations
for i in range(len(lst)):
    lst[i] = list(lst[i])
    lst[i].sort()
    lst[i] = frozenset(lst[i])
    
lst = frozenset(lst)
lst = list(lst)
for i in range(len(lst)):
    lst[i] = list(lst[i])
```

```python
# Load input data in pandas dataframe 
df = pd.read_excel('2019.02.13_repers_id.xlsx')
for i in range(len(df)):
    df.Centers_region[i] = df.Centers_region[i].lower()
    
# scan area resolution value
LM.scan_area_resolution = 3.321 #µm/px
```

```python
# IT IS MANDATORY TO RUN THIS CELL BEFORE EACH NEW RUNNING OF THE CYCLE IN THE CELL BELOW
absolute_error_result = np.zeros((1,11))
absolute_error_result
```

```python
# CYCLE
for i in range(len(lst)):
    # create a list with ID of reference landmarks:
    lst_ctrl = lst[i]
    # create a list with ID of control landmarks:
    lst_test = []
    for k in range(11):
        if k not in lst_ctrl:
            lst_test.append(k) 

    # create DataFrames for: reference landmarks -> df_ctrl, control landmarks -> df_test
    df_ctrl = df.iloc[np.array(lst_ctrl)]
    df_test = df.iloc[np.array(lst_test)]

    a_test = df_test[df_test['Centers_region'] == 'a'].iloc[:, 0:8]
    a_work = df_ctrl[df_ctrl['Centers_region'] == 'a'].iloc[:, 0:8]    
    b_test = df_test[df_test['Centers_region'] == 'b'].iloc[:, 0:8]
    b_work = df_ctrl[df_ctrl['Centers_region'] == 'b'].iloc[:, 0:8]
    c_test = df_test[df_test['Centers_region'] == 'c'].iloc[:, 0:8]
    c_work = df_ctrl[df_ctrl['Centers_region'] == 'c'].iloc[:, 0:8]
    d_test = df_test[df_test['Centers_region'] == 'd'].iloc[:, 0:8]
    d_work = df_ctrl[df_ctrl['Centers_region'] == 'd'].iloc[:, 0:8]
    e_test = df_test[df_test['Centers_region'] == 'e'].iloc[:, 0:8]
    e_work = df_ctrl[df_ctrl['Centers_region'] == 'e'].iloc[:, 0:8]
    f_test = df_test[df_test['Centers_region'] == 'f'].iloc[:, 0:8]
    f_work = df_ctrl[df_ctrl['Centers_region'] == 'f'].iloc[:, 0:8]
    g_test = df_test[df_test['Centers_region'] == 'g'].iloc[:, 0:8]
    g_work = df_ctrl[df_ctrl['Centers_region'] == 'g'].iloc[:, 0:8]

    if len(a_work) > 0:
        repers_work_a = np.array(a_work.values[:, 3:], dtype=float)
        repers_work_a = np.column_stack((repers_work_a, LM.get_stage_coord_mm(repers_work_a[:,1:], 
                                                                              a_work.values[0, 1],
                                                                              a_work.values[0, 2])))
    if len(a_test) > 0:
        repers_test_a = np.array(a_test.values[:, 3:], dtype=float)
        repers_test_a = np.column_stack((repers_test_a, LM.get_stage_coord_mm(repers_test_a[:,1:],
                                                                              a_test.values[0, 1],
                                                                              a_test.values[0, 2])))
    if len(b_work) > 0:
        repers_work_b = np.array(b_work.values[:, 3:], dtype=float)
        repers_work_b = np.column_stack((repers_work_b, LM.get_stage_coord_mm(repers_work_b[:,1:], 
                                                                              b_work.values[0, 1],
                                                                              b_work.values[0, 2])))
    if len(b_test) > 0:
        repers_test_b = np.array(b_test.values[:, 3:], dtype=float)
        repers_test_b = np.column_stack((repers_test_b, LM.get_stage_coord_mm(repers_test_b[:,1:],
                                                                              b_test.values[0, 1],
                                                                              b_test.values[0, 2])))
    if len(c_work) > 0:
        repers_work_c = np.array(c_work.values[:, 3:], dtype=float)
        repers_work_c = np.column_stack((repers_work_c, LM.get_stage_coord_mm(repers_work_c[:,1:], 
                                                                              c_work.values[0, 1],
                                                                              c_work.values[0, 2])))
    if len(c_test) > 0:
        repers_test_c = np.array(c_test.values[:, 3:], dtype=float)
        repers_test_c = np.column_stack((repers_test_c, LM.get_stage_coord_mm(repers_test_c[:,1:],
                                                                              c_test.values[0, 1],
                                                                              c_test.values[0, 2])))
    if len(d_work) > 0:
        repers_work_d = np.array(d_work.values[:, 3:], dtype=float)
        repers_work_d = np.column_stack((repers_work_d, LM.get_stage_coord_mm(repers_work_d[:,1:], 
                                                                              d_work.values[0, 1],
                                                                              d_work.values[0, 2])))
    if len(d_test) > 0:
        repers_test_d = np.array(d_test.values[:, 3:], dtype=float)
        repers_test_d = np.column_stack((repers_test_d, LM.get_stage_coord_mm(repers_test_d[:,1:],
                                                                              d_test.values[0, 1],
                                                                              d_test.values[0, 2])))
    if len(e_work) > 0:
        repers_work_e = np.array(e_work.values[:, 3:], dtype=float)
        repers_work_e = np.column_stack((repers_work_e, LM.get_stage_coord_mm(repers_work_e[:,1:], 
                                                                              e_work.values[0, 1],
                                                                              e_work.values[0, 2])))
    if len(e_test) > 0:
        repers_test_e = np.array(e_test.values[:, 3:], dtype=float)
        repers_test_e = np.column_stack((repers_test_e, LM.get_stage_coord_mm(repers_test_e[:,1:],
                                                                              e_test.values[0, 1],
                                                                              e_test.values[0, 2])))
    if len(f_work) > 0:
        repers_work_f = np.array(f_work.values[:, 3:], dtype=float)
        repers_work_f = np.column_stack((repers_work_f, LM.get_stage_coord_mm(repers_work_f[:,1:], 
                                                                              f_work.values[0, 1],
                                                                              f_work.values[0, 2])))
    if len(f_test) > 0:
        repers_test_f = np.array(f_test.values[:, 3:], dtype=float)
        repers_test_f = np.column_stack((repers_test_f, LM.get_stage_coord_mm(repers_test_f[:,1:],
                                                                              f_test.values[0, 1],
                                                                              f_test.values[0, 2])))
    if len(g_work) > 0:
        repers_work_g = np.array(g_work.values[:, 3:], dtype=float)
        repers_work_g = np.column_stack((repers_work_g, LM.get_stage_coord_mm(repers_work_g[:,1:], 
                                                                              g_work.values[0, 1],
                                                                              g_work.values[0, 2])))
    if len(g_test) > 0:
        repers_test_g = np.array(g_test.values[:, 3:], dtype=float)
        repers_test_g = np.column_stack((repers_test_g, LM.get_stage_coord_mm(repers_test_g[:,1:],
                                                                              g_test.values[0, 1],
                                                                              g_test.values[0, 2])))

    ################# normalization of reference landmarks ###########################################  
    
    repers_work = np.zeros((1,7))
    if len(a_work) > 0:
        repers_work = np.concatenate((repers_work, repers_work_a))
    if len(b_work) > 0:
        repers_work = np.concatenate((repers_work, repers_work_b))
    if len(c_work) > 0:
        repers_work = np.concatenate((repers_work, repers_work_c))
    if len(d_work) > 0:
        repers_work = np.concatenate((repers_work, repers_work_d))
    if len(e_work) > 0:
        repers_work = np.concatenate((repers_work, repers_work_e))
    if len(f_work) > 0:
        repers_work = np.concatenate((repers_work, repers_work_f))
    if len(g_work) > 0:
        repers_work = np.concatenate((repers_work, repers_work_g))
    repers_work = repers_work[1:,:]
    # mean values
    x_mean_st = repers_work[:, 5].sum() / len(repers_work)
    y_mean_st = repers_work[:, 6].sum() / len(repers_work)
    x_mean_an = repers_work[:, 3].sum() / len(repers_work)
    y_mean_an = repers_work[:, 4].sum() / len(repers_work)
    # standard deviations
    x_std_st = repers_work[:, 5].std(ddof=1)
    y_std_st = repers_work[:, 6].std(ddof=1)
    x_std_an = repers_work[:, 3].std(ddof=1)
    y_std_an = repers_work[:, 4].std(ddof=1)
    # normalization
    x_st_norm = (repers_work[:, 5] - x_mean_st) / x_std_st
    y_st_norm = (repers_work[:, 6] - y_mean_st) / y_std_st
    x_an_norm = (repers_work[:, 3] - x_mean_an) / x_std_an
    y_an_norm = (repers_work[:, 4] - y_mean_an) / y_std_an
    # matrix of basis functions
    x2 = x_st_norm**2
    y2 = y_st_norm**2
    xy = x_st_norm * y_st_norm
    # x = x_st_norm
    # y = y_st_norm
    ones = np.ones(len(repers_work))
    matrix_of_basic_functions = np.column_stack((x2, y2, xy, x_st_norm, y_st_norm, ones))
    # transposed matrix of basis functions
    transposed_MBF = matrix_of_basic_functions.T
    # matrix of normal equations (ХTХ)
    matrix_of_normal_equations = np.dot(transposed_MBF, matrix_of_basic_functions)

    # Fisher matrix  (ХTХ)^-1
    Fisher_matrix = np.linalg.inv(matrix_of_normal_equations)
    # Fisher matrix * transposed matrix of basis functions)
    Fisher_x_transposed = np.dot(Fisher_matrix, transposed_MBF)
    # coeffcients vectors
    coefficients_vector_x = np.dot(Fisher_x_transposed, x_an_norm)
    coefficients_vector_y = np.dot(Fisher_x_transposed, y_an_norm)
    
    ################# normalization of control landmarks ########################################### 

    repers_test = np.zeros((1,7))
    if len(a_test) > 0:
        repers_test = np.concatenate((repers_test, repers_test_a))
    if len(b_test) > 0:
        repers_test = np.concatenate((repers_test, repers_test_b))
    if len(c_test) > 0:
        repers_test = np.concatenate((repers_test, repers_test_c))
    if len(d_test) > 0:
        repers_test = np.concatenate((repers_test, repers_test_d))
    if len(e_test) > 0:
        repers_test = np.concatenate((repers_test, repers_test_e))
    if len(f_test) > 0:
        repers_test = np.concatenate((repers_test, repers_test_f))
    if len(g_test) > 0:
        repers_test = np.concatenate((repers_test, repers_test_g))
    repers_test = repers_test[1:, :]

    # normalization
    x_st_norm_t = (repers_test[:, 5] - x_mean_st) / x_std_st
    y_st_norm_t = (repers_test[:, 6] - y_mean_st) / y_std_st
    # matrix of basis functions for control landmarks
    x2_t = x_st_norm_t**2
    y2_t = y_st_norm_t**2
    xy_t = x_st_norm_t * y_st_norm_t
    # x = x_st_norm_t
    # y = y_st_norm_t
    ones_t = np.ones(len(repers_test))
    matrix_of_basic_functions_t = np.column_stack((x2_t, y2_t, xy_t, x_st_norm_t, y_st_norm_t, ones_t))
    x_an_norm_t = np.dot(matrix_of_basic_functions_t, coefficients_vector_x)
    y_an_norm_t = np.dot(matrix_of_basic_functions_t, coefficients_vector_y)
    # get anatomical coordinations of control landmarks
    x_an_t =  x_an_norm_t * x_std_an + x_mean_an
    y_an_t =  y_an_norm_t * y_std_an + y_mean_an
    # error of calculations
    abs_error_x = (x_an_t - repers_test[:, 3])
    abs_error_y = (y_an_t - repers_test[:, 4])
  
    result = np.column_stack((repers_test, x_an_t, y_an_t, abs_error_x, abs_error_y))

    absolute_error_result = np.concatenate((absolute_error_result, result))

    if len(a_work) > 0:
        del repers_work_a
    if len(a_test) > 0:
        del repers_test_a
    if len(b_work) > 0:
        del repers_work_b
    if len(b_test) > 0:
        del repers_test_b
    if len(c_work) > 0:
        del repers_work_c
    if len(c_test) > 0:
        del repers_test_c
    if len(d_work) > 0:
        del repers_work_d
    if len(d_test) > 0:
        del repers_test_d
    if len(e_work) > 0:
        del repers_work_e
    if len(e_test) > 0:
        del repers_test_e
    if len(f_work) > 0:
        del repers_work_f
    if len(f_test) > 0:
        del repers_test_f
    if len(g_work) > 0:
        del repers_work_g
    if len(g_test) > 0:
        del repers_test_g
```

```python
absolute_error_result = absolute_error_result[1:,:]
error_df = pd.DataFrame(data=absolute_error_result, columns=['id_point', '2', '3', '4', '5', '6', '7', '8', '1', 'X', 'Y'])
error_df = error_df.drop(columns={'2', '3', '4', '5', '6', '7', '8', '1'})
```

```python
error_df['Error'] = (error_df.X ** 2 + error_df.Y ** 2)**0.5
```

```python
# save results of calculations of errors into a file 
error_df.to_excel('MSE_error.xlsx', index=False)
```
